```python
import tensorflow as tf
```

```python
daily_sales = [21,22,-108,31,-1,32,34,31]
```

```python
tf_dataset = tf.data.Dataset.from_tensor_slices(daily_sales)
tf_dataset
```

```python
for sales in tf_dataset:
    print(sales)
```

```python
# to convert into a numpy object
for sales in tf_dataset:
    print(sales.numpy())
```

```python
# or use this other method to convert into a numpy object
for sales in tf_dataset.as_numpy_iterator():
    print(sales)
```

```python
# to get the first 3 elements
for sales in tf_dataset.take(3):
    print(sales.numpy())
```

```python
# to filter invalid data points
tf_dataset = tf_dataset.filter(lambda x: x>0)
for sales in tf_dataset.as_numpy_iterator():
    print(sales)
```

```python
# covert to fcfa
tf_dataset = tf_dataset.map(lambda x: x*650)
for sales in tf_dataset.as_numpy_iterator():
    print(sales)
```

```python
# shuffle
tf_dataset = tf_dataset.shuffle(2)
for sales in tf_dataset.as_numpy_iterator():
    print(sales)
```

```python
# create batches
for sales in tf_dataset.batch(2):
    print(sales.numpy())
```

```python
# write one function that discribes all these steps above
tf_dataset = tf.data.Dataset.from_tensor_slices(daily_sales)
tf_dataset = tf_dataset.filter(lambda x: x>0).map(lambda y: y*650).shuffle(2).batch(2)
for sales in tf_dataset.as_numpy_iterator():
    print(sales)
```
