```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Histogram and Distribution visualizations

```python
#Histogram Big Game (data)
data.hist(bins = 20, figsize=(12, 9))
plt.show()
```

### statistics 
to understand the average sales, the range of sales, and the variation in sales figures.

```python
# Creating Sales colunm
data['Sales'] = data['Payment (5)'] 
```

```python
print(data.head())
```

```python
# Creating Average sales, range of sales & variationnn in sales figures.
average_sales = data['Sales'].mean()
sales_range = data['Sales'].max() - data['Sales'].min()
variance_sales = data['Sales'].var()
```

```python
print(f"Average Sales: {average_sales}")
print(f"Range of Sales: {sales_range}")
print(f"Variance of Sales: {variance_sales}")
```

### Correlationn Analysis

```python
data = data.drop(columns=['Licensee', 'Fiscal Year', 'Month Ending', 'Sales'])
```

```python
data.head()
```

```python
# Correlation Matrix

corr_matrix = data.corr()
```

```python
corr_matrix
```

```python
plt.figure(figsize=(10, 6))
plot = sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

#  bar charts and scatter plots
### bar charts

```python
file = 'Selected_Online_Casino_Gaming_Data.csv'
data = pd.read_csv(file)
data['Sales'] = data['Payment (5)'] 
```

```python
data.head()
```

```python
# bar chart for Licensee
licensee = data['Licensee']
sales = data['Sales']
```

```python
fig = plt.figure(figsize=(15, 10))
plt.bar(licensee, sales, color='olive')
plt.xlabel('licensee')
plt.ylabel('sales')
plt.title('Licensee')
plt.show()
```

```python
# bar chart for Fiscal Year
fiscal_year = data['Fiscal Year']
sales = data['Sales']
```

```python
fig = plt.figure(figsize=(15, 10))
plt.bar(fiscal_year, sales, color='navy')
plt.xlabel('fiscal_year')
plt.ylabel('sales')
plt.title('Fiscal Year')
plt.show()
```

```python
# bar chart for Month Ending
month_ending = data['Month Ending']
sales = data['Sales']
```

```python
fig = plt.figure(figsize=(15, 10))
plt.bar(month_ending, sales, color='maroon')
plt.xlabel('month_ending')
plt.ylabel('sales')
plt.title('Month Ending')
plt.show()
```

### Scatter plots

```python
# Scatter plot for Licensee
x = licensee
y = sales


fig = plt.figure(figsize=(15, 10))
plt.scatter(x,y)
plt.colorbar(label = 'Color Values')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot for Licensee')
plt.show()
```

```python
# Scatter plot for Fiscal Year
x = fiscal_year
y = sales

fig = plt.figure(figsize=(15, 10))
plt.scatter(x,y)
plt.colorbar(label = 'Color Values')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot for Fiscal Year')
plt.show()
```

```python
# Scatter plot for Month Ending
x = month_ending
y = sales

fig = plt.figure(figsize=(15, 10))
plt.scatter(x,y)
plt.colorbar(label = 'Color Values')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot for Month Ending')
plt.show()
```
